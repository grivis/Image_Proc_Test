## Splitting Images into Colours

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
sunflower = imageio.imread('sunflower_1.jpg')
```

```python
plt.imshow(sunflower)
```

```python
sunflower.shape
```

```python
red = sunflower[:,:,0]
green = sunflower[:,:,1]
blue = sunflower[:,:,2]
```

```python
red.shape
```

```python
## The first two dimensions are the same. That means we haven't changed the picture's geometry
```

```python
green.shape
```

```python
blue
```

```python
### What we get if we try to display the images individually?
```

```python
plt.imshow(red, cmap='gray')
```

```python
## The above is the original picture in RED
```

```python
## White spots are spots of the original picture with a lot of red...
## Black spots are spots with almost no red
```

```python
plt.imshow(green, cmap='gray')
```

```python
plt.imshow(blue, cmap='gray')
```

```python
## The sky is blue and that is why very light in blue picture
## The flower itself has very little blue and looks almost black
```

```python
## The colours must follow the exact order: RED, GREEN, BLUE
```

```python
## What if we put them in a wrong order? e.g. BLUE, GREEN, RED
```

```python
## Restore the 3D array... But with wrong colours
```

```python
## Combine 3 2D arrays into a single 3D array
```

```python
## Merge 2 arrays...
```

```python
canvas = np.zeros((1329, 1200, 3), dtype='uint8')
```

```python
plt.imshow(canvas)
```

```python
canvas[:,:,0] = blue
canvas[:,:, 1] = red
canvas[:,:, 2] = green
```

```python
plt.imshow(canvas)
```

```python
## The above picture shows what happens if we confuse colours...
```

```python
## To demonstrate the correct order, merge colours in the RIGHT way
```

```python
canvas[:,:,0] = red
canvas[:,:, 1] = green
canvas[:,:, 2] = blue
```

```python
plt.imshow(canvas)
```

### Converting Colour Picture into Black and White

```python
'''
What is the difference between a colour image and a monochrome (bW) image?

'''
```

```python
sf_bw = imageio.imread('sunflower_bw.jpg')
```

```python
plt.imshow(sf_bw, cmap='gray')
```

```python
sf_bw.shape
```

```python
### BW is always 2D array!
```

```python
## NO colour layers!!
```

```python
sf_bw[:10, :10]
```

```python
## Every pixel of the picture holds only ONE value (not three!)

## this value represents shades of grey 0...255
```

```python
### A HINT: to convert colour image into BW, you need to make a single value out if three
## R, G, B >>>> BW
```
